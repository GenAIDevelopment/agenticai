```python
from pypdf import PdfReader

def extract_images_from_pdf(pdf_path, output_folder="."):
    """
    Extracts all images from a given PDF file and saves them to an output folder.
    """
    reader = PdfReader(pdf_path)
    print(f"Opened PDF: {pdf_path}")
    image_count = 0

    for page_num, page in enumerate(reader.pages):
        print(f"Processing page {page_num + 1}...")
        for i, image_file_object in enumerate(page.images):
            # Generate a unique file name
            file_name = f"{output_folder}/page{page_num+1}-image{i+1}-{image_file_object.name}"
            
            try:
                # Save the image data
                with open(file_name, "wb") as f:
                    f.write(image_file_object.data)
                print(f"  Saved image: {file_name}")
                image_count += 1
            except Exception as e:
                print(f"  Error saving image {file_name}: {e}")

    print(f"Extraction complete. Total images saved: {image_count}")

# Example usage:
# Replace "example.pdf" with the path to your PDF file
extract_images_from_pdf(ncert_book_path, output_folder="extracted_images")
```

# Experimenting with passing multi modal 

```python
from langchain.messages import HumanMessage

# image is hosted publically

url = "https://cdn-fckad.nitrocdn.com/QQqEhdxzQCfGsIbVrVWwmnFZeiIEPRRw/assets/images/optimized/rev-d1768af/windshield-expert.s3.ap-south-1.amazonaws.com/wp-content/uploads/2024/06/14084059/Best-5-Tata-Cars.jpg"
 
human_message = HumanMessage(content=[
    {"type": "text", "text": "Extract the information from the image"},
    {"type": "image_url", "image_url": {"url": url}}
])
```

```python
from langchain_google_vertexai import ChatVertexAI
```

```python
model = ChatVertexAI(
    model_name="gemini-2.5-flash",
)
```

```python
response = model.invoke([human_message])
```

```python
response.pretty_print()
```

```python
# image is in local file system
import base64

# When the images are local we need to embedd the images in prompt as text
def image_to_data_url(path):
    """Converts an image file to a data URL."""
    with open(path, "rb") as image_file:
        encoded_image = base64.b64encode(image_file.read()).decode("utf-8")
        # This works only with jpeg
        return f"data:image/jpeg;base64,{encoded_image}"
```

```python
image_path = 'panchatantra_images/page1-image1-R13.jpg'
image_data = image_to_data_url(image_path)
```

```python
human_message = HumanMessage(content=[
    {"type": "text", "text": "Extract the information from the image"},
    {"type": "image_url", "image_url": {"url": image_data}}
])
```

```python
response = model.invoke([human_message])
```

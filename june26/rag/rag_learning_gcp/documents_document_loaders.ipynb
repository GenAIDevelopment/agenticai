```python
story="""Once upon a time, a quiet clockmaker named Thomas lived in a small valley surrounded by foggy hills. He spent his days in a dusty shop, surrounded by ticking clocks of all shapes and sizes. Every gear he turned and every spring he wound brought him peace. The townspeople brought him their broken timepieces, knowing his hands could fix anything. Yet, Thomas always felt he was waiting for something unusual to happen.

One chilly evening, a cloaked traveler entered his shop carrying a heavy wooden box. The stranger placed it on the counter, whispered that it could not be fixed by ordinary means, and vanished into the mist. Inside the box lay a strange brass device with glowing runes instead of numbers. It did not tick; it hummed with a low, vibrating energy that filled the room. Thomas felt a thrill of excitement he had not experienced in years.

He stayed up all night examining the mysterious mechanism under his magnifying glass. As his brass tools touched the main gear, the room fell completely silent, and the ticking of his other clocks stopped. The air grew warm, and the glowing runes began to spin backwards at high speed. Outside his window, the falling rain froze mid-air, suspended like tiny diamonds in the dark night. He had accidentally paused time itself.

Realizing the immense power in his hands, Thomas carefully turned the gear back to its original position. The clocks immediately resumed their ticking, and the rain splashed loudly against the glass. He knew this device was too dangerous to be left unattended or forgotten. He closed his shop, packed his tools into a leather bag, and set off into the foggy hills to find the traveler and uncover the truth.
"""
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100,
    separators=["."]
)
chunks = splitter.split_text(
    story
)
```

```python
len(chunks)
```

```python
for chunk in chunks:
    print(chunk)
```

```python
# doc structure based chunking

from langchain_text_splitters.markdown import MarkdownHeaderTextSplitter

md_doc = TextLoader(
    file_path="./data/story.md",
    encoding="utf-8"
)

splitter = MarkdownHeaderTextSplitter(headers_to_split_on=[
    ("#", "Header 1"),
    ("##", "Header 2"),
    ("###", "Header 3")])
```

```python
chunks = splitter.split_text(
    md_doc.load()[0].page_content)
```

```python
len(chunks)
```

```python
print(chunks[0].page_content)
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

loader = TextLoader(
    file_path="./data/story.txt",
    autodetect_encoding=True
)

splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20
)

# chunks
documents = loader.load_and_split(splitter)
```

```python
len(documents)
```

```python
for document in documents:
    print(document.page_content)
    print(document.metadata)
    
```

## Prompt Templating

```python
from utils import get_model_from_gcp
llm = get_model_from_gcp()
```

```python
# I want to prompt to llm find a capital city of dynamic country

# python based approach
country = input("Enter any country name: ")
prompt = f"What is capital of {country}"
response = llm.invoke(prompt)
response.pretty_print()
```

```python
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate.from_template(
    template = "What is capital of {country}? and Also show their acheivements in the following sport {sport}",
    )

chain = prompt_template | llm

response = chain.invoke({
    "country": "India",
    "sport": "Hockey"
})
print(response)

response = chain.invoke({
    "country": "Portugal",
    "sport": "Football"
})
response
```

```python
from langchain_core.prompts import ChatPromptTemplate

# prompt with roles

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an expert in {topic}."),
    ("user", "What is significance of {topic} for {audience}")
]
)

prompt.invoke({
    "topic": "Sports",
    "audience": "students"
})
```

```python
chain = prompt | llm

response = chain.invoke({
    "topic": "Sports",
    "audience": "students"
})
print(response.content)
```

## Output structure

* When we get response from llm we want outputs in a specific structure



```python
from langchain_core.output_parsers import StrOutputParser
output = StrOutputParser()
chain = prompt | llm | output

response = chain.invoke({
    "topic": "Sports",
    "audience": "students"
})
print(response)
```

```python
from pydantic import BaseModel, Field


class SupportTicket(BaseModel):
    issue_category: str = Field(description="One of: billing, technical, shipping, other")
    resolved: bool = Field(description="Whether the issue appears resolved based on the conversation")
    next_action: str = Field(description="What should happen next, one short sentence")


prompt = ChatPromptTemplate.from_messages([
    ("system", "you summarize customer support conversations into structured tickets."),
    ("user", "My last order never arrived and it's been 2 weeks"),
    ("assistant", "I'm sorry to hear that, I have issued a reshipment, tracking number sent to your registred email and phone number"),
    ("user", "Got it thank you")
])

chain = prompt | llm.with_structured_output(SupportTicket, method='json_schema')
response = chain.invoke({})
response
```

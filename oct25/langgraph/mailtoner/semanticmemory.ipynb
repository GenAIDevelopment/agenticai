```python
import os
from langgraph.graph import StateGraph, MessagesState, START, END
from langgraph.store.postgres import PostgresStore
from dotenv import load_dotenv
```

```python
from utils import get_llm

model = get_llm()
```

```python
load_dotenv()
```

```python
from langchain_google_vertexai.embeddings import VertexAIEmbeddings

def get_embedding(name:str = 'text-embedding-005') -> VertexAIEmbeddings:
    return VertexAIEmbeddings(model=name)

def ns(user_id:str) -> tuple:
    return ("memories", user_id)
```

```python
DB_URI = os.getenv('PG_VECTOR_DB_URI')

with PostgresStore.from_conn_string(
    DB_URI, 
    index={
            "dims": 768,
                "embed": get_embedding(),
                "fields": ["memory_text"]  # specify which fields to embed. Default is the whole serialized value
            }
    ) as store:
    store.setup()
```

```python
import uuid
user_id = "dummy"
with PostgresStore.from_conn_string(
    DB_URI, 
    index={
            "dims": 768,
                "embed": get_embedding(),
                "fields": ["memory_text"] # specify which fields to embed. Default is the whole serialized value
            }
    ) as store:
    store.put(
        namespace=ns(user_id),
        key = str(uuid.uuid4()),
        value={"memory_text": "I Study computers"}

    )

    store.put(
        namespace=ns(user_id),
        key = str(uuid.uuid4()),
        value={"memory_text": "I like pizza's"}

    )
    store.put(
        namespace=ns(user_id),
        key = str(uuid.uuid4()),
        value={"memory_text": "I like tea"}

    )

    store.put(
        namespace=ns(user_id),
        key = str(uuid.uuid4()),
        value={"memory_text": "I like biryani"}
    )

    store.put(
        namespace=ns(user_id),
        key = str(uuid.uuid4()),
        value={"memory_text": "I like studying opensource code "}
    )

    store.put(
        namespace=ns(user_id),
        key = str(uuid.uuid4()),
        value={"memory_text": "Programming is a major area of intrest"}
    )

```

```python
user_id = "dummy"
with PostgresStore.from_conn_string(
    DB_URI, 
    index={
            "dims": 768,
                "embed": get_embedding(),
                "fields": ["memory_text"] # specify which fields to embed. Default is the whole serialized value
            }
    ) as store:
    hits = store.search(
        ns(user_id),
        query="what do i like to eat",
        limit=2
    )
    for hit in hits:
        print(f"score: {hit.score}, result={hit.value['memory_text']}")
```
